```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.dropna(inplace=True)
```

```python
#now we wish to split the data into training and testing data, so we will make the required demaarcation
from sklearn.model_selection import train_test_split

X=data.drop(['median_house_value'], axis=1)
y=data['median_house_value']
```

```python
#we split the data and after splitting it, have a seperate train data to train the machine learning model and the test data to test how well the model has trained 
#so for this, we have kept aside 20% of the data for testing purpose exclusively 
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2)
```

```python
train_data= X_train.join(y_train)
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(train_data.corr(method='pearson', numeric_only=True), annot=True, cmap="YlGnBu")
```

```python
#data exploration is doen to understand the basic essence of the data using various representation techniques 
train_data['total_rooms']=np.log(train_data['total_rooms'] +1)
train_data['total_bedrooms']=np.log(train_data['total_bedrooms'] +1)
train_data['population']=np.log(train_data['population'] +1)
train_data['households']=np.log(train_data['households'] +1)
```

```python
train_data.hist(figsize=(16,8))
```

```python
#pre process the ocean proximity data value count
train_data.ocean_proximity.value_counts()
```

```python
# Perform one-hot encoding
one_hot_encoded = pd.get_dummies(train_data['ocean_proximity'], sparse=True, dtype=float)
print(one_hot_encoded)
```

```python
train_data=train_data.join(one_hot_encoded).drop(['ocean_proximity'], axis=1)
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(train_data.corr(method='pearson', numeric_only=True), annot=True, cmap="YlGnBu")
```

```python
#though this is not a map of california the plot makes it almost like the map and gives us an idea in data representation
plt.figure(figsize=(16,9))
sns.scatterplot(x="latitude", y="longitude", data=train_data, hue="median_house_value", palette="coolwarm")
```

```python
train_data['bedroom_ratio']=train_data['total_bedrooms']/ train_data['total_rooms']
train_data['household_rooms']=train_data['total_rooms']/ train_data['households']
```
